# Aufgabe 3 a


Betrachte 
$$
f(z)= \overline{z}^n= (x-iy)^n = \sum_{k=0}^n \binom{n}{k} x^{n-k} (-iy)^n
$$
Teilt man $f$ auf in Realteil und Imaginärteil auf, so ergibt sich für reelle Funktionen $u,v$, wobei $f = u + iv$, folgendes:
$$
\begin{align*}
u(z) & = \sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k} x^{n-2k} (-iy)^{2k} = \\
& =\sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k} x^{n-2k}  y^{2k} (-1)^k \\\\

v(z) & = -i \sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k+1} x^{n-2k-1} (-iy)^{2k+1} = \\
& =- \sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k+1} x^{n-2k-1}  y^{2k+1} (-1)^k
\end{align*}
$$

Betrachtet man die partiellen Ableitungen,

\begin{align*}
\frac{\partial u }{\partial x}(z) & =\;\;\;\sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k}\qquad (n-2k)\; x^{n-2k-1}  y^{2k} (-1)^k \\\\

\frac{\partial v }{\partial y}(z) & =- \sum_{k=0}^{\lfloor\frac{n}{2}\rfloor} \binom{n}{2k+1} (2k+1)\; x^{n-2k-1}  y^{2k} (-1)^k

\end{align*}

so fällt durch einen Koeffizientenvergleich auf, dass im Allgemeinen $\frac{\partial u }{\partial x}(z) \neq \frac{\partial v }{\partial y}(z)$. Somit gelten nicht die Reimanschen Differentialgleichungen, obwohl $u,v$ durchaus stetig differenzierbar sind.
Daraus folgt, dass $f$ für alle 
$n$ > 0 nicht holomorph ist. Ist nun $n=0$, so ist $f$ konstant und trivialerweise holomorph.

$$ 
\begin{align*}
U_{AB} & = \{z \in \mathbb{C}: c_{AB} \cdot z + d_{AB} \neq 0\}\\

$$

Betracthe nun $x \in M_B^{-1}(U_A) \iff M_B(x) \in U_A$.

$$c_2 x + d_2 \neq 0$$

$$
\begin{align*}
c_1 M_B(x) + d_1 &\neq 0\\

c_1 (a_2 x + b_2) + d_1 \cdot({c_2 x + d_2}) &\neq 0\\

\end{align*}
$$

Somit folgt also aus $x \in M_B^{-1}(U_A)$ direkt $x \in U_{AB}$.\\
Und dies impliziert $M_B^{-1}(U_A) \subset U_{AB}$.

Nun soll noch gezeigt werden, dass $M_A \circ M_B = M_{AB}$ auf $M_B^{-1}(U_A)$ ist. Sei also $z \in M_B^{-1}(U_A)$, dann

## c)
Im Folgendem bezeichne $C = \mathbb{C} \cup \{\infty\}$ und $M = M_A$

### Fall $c=0$
Betrachte zunächst nur den Fall $c=0$. Dann muss $d\neq 0$, denn andernfalls ist $A$ nicht invertierbar. Also sind $M_A$ und $U_A$ gegeben durch
$$ 
U:= U_A =\mathbb{C} \\ 
M_A = \frac{a}{d} z + b
$$

#### Natürliche Fortsetzung

Somit kann man $M_A$ natürlich erweitern auf $C$ mit $\hat{M_A}(\infty) = \infty$.

Um zu zeigen, dass dies eine Bijektion auf $C$ darstellt, reicht es zu zeigen, dass $M_A$ bereits auf $\mathbb{C}$ eine Bijektion darstellt. Dies liegt daran, dass $M_A(x) \neq \infty$ für $x \in \mathbb{C}$.

#### injektiv
Seien $x,y \in \mathbb{C}$ mit $M(x) = M(y)$. Dann ist

$$ 
\begin{align*}
\frac{a}{d} x + b  = & \frac{a}{d} y + b\\
\frac{a}{d} x = & \frac{a}{d} y \\
x = &  y \\
\end{align*}
$$
Also ist $M = M_A$ injektiv.

##### surjektiv
Sei $y \in \mathbb{C}$. Dann gilt für $x = \frac{d}{a} (y - b)$: 
$$ M(x) = y-b +b = y$$ 
Also ist $M = M_A$ surjektiv.





### Fall $c\neq0$
Betrachte
$$ M(z) = \frac{a z +b}{cz + d} = \frac{a+b/z}{c + d/z}$$
Somit ergibt sich die natürliche Erweiterung durch:
$$
M(\infty) = a/c\\
M(-d/c) = \infty
$$ 
Auch hier reicht es zu zeigen, dass M auf $U_AB$ bereits eine Bijektion auf ihr Bild dargestellt hat.

#### injektiv
Seien  $x,y \in \mathbb{C}$ mit $M(x) = M(y)$. Dann ist

$$ 
\begin{align*}
\frac{a x +b}{cx + d} = &\frac{a y +b}{cy + d}\\
(ax + b)(cy +d) =& (ay+b)(cx +d )\\
ac\cdot xy + bd + ad \cdot x + bc \cdot y = &  ac \cdot xy + bd + ad \cdot y + bc \cdot x\\
ad \cdot x + bc \cdot y = &  ad \cdot y + bc \cdot x\\
(ad-bc) \cdot x  = &  (ad-bc) \cdot y \\
x =y 
\end{align*}
$$
Der letzte Schritt ist durchführbar, da $0 \neq det(A) = ad - bc$, weil $A$ invertierbar.




#### surjektiv
Wähle für ein $x\in U_A \setminus \{\frac{c}{a}\}$:
$$ 
z = \frac{dx - b}{a-cx}
$$
Dann gilt
$$
\begin{align*}
M(z) &= \frac{a \cdot \frac{dx - b}{a-cx} + b}{c \cdot \frac{dx - b}{a-cx} + d}\\
&=\frac{a \cdot (dx - b) + (a-cx) \cdot b}{c \cdot (dx - b) + (a-cx) \cdot d}\\
&=\frac{ad\cdot x - cb \cdot x }{c \cdot (dx - b) + (a-cx) \cdot d}\\
&=\frac{ad\cdot x - cb \cdot x }{ad - cb} = x\\

\end{align*}
$$
Somit ist es surjektiv auf das Bil.
